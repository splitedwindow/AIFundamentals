***Завдання 5:*** В моделях машинного навчання інснує велика проблема опрацювання безперервних змінних. Потрібно перетворити безперервні значення в категоріальні. 

Максимальний вік пасажира - ```80 років```.

Отже, давайте розділимо діапазон від ```0 до 80``` на ```5``` осередків. 


Додайте до нашого датасету нову колонку ```Age_bin```, в якій будуть значення ```0, 1, 2, 3, 4```відповідно кожній віковій групі.

Побудуйте графік, який відображає кількість виживших у кожній віковій категорії.





````python
median_age = dataset['Age'].median()
dataset['Age'].fillna(median_age, inplace=True)

intervals = [0, 16, 32, 48, 64, 81]

categories = ['0-15', '16-31', '32-47', '48-63', '64-80']

dataset['Age_Category'] = pd.cut(dataset['Age'], bins=intervals, labels=categories, right=False)
````

````python
survived_categories = []

for category in categories:
    survived_in_category = dataset.loc[(dataset['Age_Category'] == category) & (dataset['Survived'] == 1), 'Survived'].sum()
    survived_categories.append(survived_in_category)

plt.figure(figsize=(5, 15))
plt.bar(categories, 
        survived_categories,
        label='Survived',
       zorder=3)
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.title('Count of survived passanger within each age category.')

plt.legend()
plt.grid(zorder=0)
plt.show() 
````

Кількість виживших пасажирів в кожній віковій категорії по 16 років.

Корисних висновків на основі лише кількості людей зробити не можна, адже кількість людей в кожній категорії відрізняється. Цікавіші висновки можна було б зробити на основі відсотку виживших в кожній з категорій. 

***Завдання 6:*** Так як змінна ```fare``` також є неперервною її треба закодувати до ```порядкової шкали``` 
Для цього скоритайтесь функцією ```pandas.qcut```.
[pandas.qcut](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.qcut.html)  
- необхідно перевірити, чи є вірною гіпотеза про те, що кількість виживших пасажирів для квартильних значеннь змінної ```fare``` відрізняється, та чи доцільно використовувати кодування на цю змінну(для подальшої роботи з закодованою фичею)

````python
import pandas as pd
from scipy.stats import chi2_contingency

dataset['Fare_quartile'] = pd.qcut(dataset['Fare'], q=4, labels=False)

contingency_table = pd.crosstab(dataset['Survived'], dataset['Fare_quartile'])

chi2, p_value, _, _ = chi2_contingency(contingency_table)

alpha = 0.05
print("Chi-square statistic:", chi2)
print("p-value:", p_value)

if p_value < alpha:
    print("There is significant evidence to reject the null hypothesis.")
    print("There is a significant association between fare quartile categories and survival status.")
else:
    print("There is no significant evidence to reject the null hypothesis.")
    print("There is no significant association between fare quartile categories and survival status.")
````

***Завдання 7:*** Необхідно створити нові змінні ```Family_size``` та ```Alone``` та проаналізувати їх. Змінна ```Family_size``` передає суму змінних ```Parch``` та ```SibSp```. Змінна ```Family_size``` буде відібражати комбіновані дані, які необхідні для перевірки гіпотези, що ```виживання пасажирів залежіть від кількості членів їх сім'ї```. Змінна ```Alone``` позначає наявніть сім'ї.

````python
dataset['Family_size'] = dataset['Parch'] + dataset['SibSp']

dataset['Alone'] = 0 
dataset.loc[dataset['Family_size'] == 0, 'Alone'] = 1

plt.figure(figsize=(10, 5))

sns.countplot(x='Family_size', hue='Survived', data=dataset)
plt.title('Survival by Family Size')

plt.figure(figsize=(6, 5))
sns.countplot(x='Alone', hue='Survived', data=dataset)
plt.title('Survival of Alone Passengers')

plt.show()
````

Людей які були самі на борту, померло більш ніж вдвічі аніж вижило. Людей які не були самі на борту загинуло та вижило однакова кількість.

виживання пасажирів в якійсь мірі залежіть від кількості членів їх сім'ї
